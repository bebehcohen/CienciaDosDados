```python
#definindo o nome para os diferentes arquivos
dados13=pd.read_csv('DOM2013.csv')
dados09=pd.read_csv('DOM2009.csv')
```

```python
#definindo as variaveis qualitativas em categorias
dados13.V0105 = dados13.V0105.astype('float') #preocupacao
dados09.V0105 = dados09.V0105.astype('float') #preocupacao


dados13.V2103 = dados13.V2103.astype('category') #preocupacao
dados13.V2103.cat.categories = (['Sim', 'Não'])
dados09.V2103 = dados09.V2103.astype('category') #preocupacao
dados09.V2103.cat.categories = (['Sim', 'Não'])


dados13.V2105 = dados13.V2105.astype('category') #alimentos acabaram
dados13.V2105.cat.categories = (['Sim', 'Não'])
dados09.V2105 = dados09.V2105.astype('category') #alimentos acabaram
dados09.V2105.cat.categories = (['Sim', 'Não'])


dados13.V2107 = dados13.V2107.astype('category') #sem dinheiro para uma alimentacao saudavel
dados13.V2107.cat.categories = (['Sim', 'Não'])
dados09.V2107 = dados09.V2107.astype('category') #sem dinheiro para uma alimentacao saudavel
dados09.V2107.cat.categories = (['Sim', 'Não'])


dados13.V2109 = dados13.V2109.astype('category') #comeram pouco pois dinheiro acabou
dados13.V2109.cat.categories = (['Sim', 'Não'])
dados09.V2109 = dados09.V2109.astype('category') #comeram pouco pois dinheiro acabou
dados09.V2109.cat.categories = (['Sim', 'Não'])


dados13.V2113 = dados13.V2113.astype('category') #deixou de comer
dados13.V2113.cat.categories = (['Sim', 'Não'])
dados09.V2113 = dados09.V2113.astype('category') #deixou de comer
dados09.V2113.cat.categories = (['Sim', 'Não'])


dados13.V2115 = dados13.V2115.astype('category') #comeu menos
dados13.V2115.cat.categories = (['Sim', 'Não'])
dados09.V2115 = dados09.V2115.astype('category') #comeu menos
dados09.V2115.cat.categories = (['Sim', 'Não'])


dados13.V2117 = dados13.V2117.astype('category') #sentiu fome1
dados13.V2117.cat.categories = (['Sim', 'Não'])
dados09.V2117 = dados09.V2117.astype('category') #sentiu fome1
dados09.V2117.cat.categories = (['Sim', 'Não'])


dados13.V2121 = dados13.V2121.astype('category') #dia inteiro sem comer1
dados13.V2121.cat.categories = (['Sim', 'Não'])
dados09.V2121 = dados09.V2121.astype('category') #dia inteiro sem comer1
dados09.V2121.cat.categories = (['Sim', 'Não'])


dados13.V2124 = dados13.V2124.astype('category') #deixou de ter alimentacao saudavel
dados13.V2124.cat.categories = (['Sim', 'Não'])
dados09.V2124 = dados09.V2124.astype('category') #deixou de ter alimentacao saudavel
dados09.V2124.cat.categories = (['Sim', 'Não'])


dados13.V2126 = dados13.V2126.astype('category') #nao comeu suficiente
dados13.V2126.cat.categories = (['Sim', 'Não'])
dados09.V2126 = dados09.V2126.astype('category') #nao comeu suficiente
dados09.V2126.cat.categories = (['Sim', 'Não'])


dados13.V2128 = dados13.V2128.astype('category') #diminuida quantidade de comida
dados13.V2128.cat.categories = (['Sim', 'Não'])
dados09.V2128 = dados09.V2128.astype('category') #diminuida quantidade de comida
dados09.V2128.cat.categories = (['Sim', 'Não'])


dados13.V2130 = dados13.V2130.astype('category') #deixou de fazer refeicao
dados13.V2130.cat.categories = (['Sim', 'Não'])
dados09.V2130 = dados09.V2130.astype('category') #deixou de fazer refeicao
dados09.V2130.cat.categories = (['Sim', 'Não'])


dados13.V2132 = dados13.V2132.astype('category') #sentiu fome2
dados13.V2132.cat.categories = (['Sim', 'Não'])
dados09.V2132 = dados09.V2132.astype('category') #sentiu fome2
dados09.V2132.cat.categories = (['Sim', 'Não'])


dados13.V2134 = dados13.V2134.astype('category') #dia inteiro sem comer2
dados13.V2134.cat.categories = (['Sim', 'Não'])
dados09.V2134 = dados09.V2134.astype('category') #dia inteiro sem comer2
dados09.V2134.cat.categories = (['Sim', 'Não'])




```

```python
#renomeando as variaveis e excluindo as NaN
d13=dados13.loc[:,['V0105', 'V2103', 'V2105', 'V2107', 'V2109', 'V2113', 'V2115', 'V2117', 'V2121', 'V2124','V2126', 'V2128', 'V2130', 'V2132', 'V2134']]
d13.columns=["Total Moradores", "Preocupação", "Alimentos acabaram", "Sem dinheiro para uma alimentação saudável", "Comeram pouco pois dinheiro acabou", "Deixou de comer","Comeu menos", "Sentiu fome1", "Dia inteiro sem comer1","Deixou de ter alimentação saudável", "Não comeu suficiente", "Diminuida quantidade de comida", "Deixou de fazer uma refeição", "Sentiu fome2", "Dia inteiro sem comer2"]
d13.dropna(inplace=True)

d09=dados09.loc[:,['V0105', 'V2103', 'V2105', 'V2107', 'V2109', 'V2113', 'V2115', 'V2117', 'V2121', 'V2124','V2126', 'V2128', 'V2130', 'V2132', 'V2134']]
d09.columns=["Total Moradores", "Preocupação", "Alimentos acabaram", "Sem dinheiro para uma alimentação saudável", "Comeram pouco pois dinheiro acabou", "Deixou de comer","Comeu menos", "Sentiu fome1", "Dia inteiro sem comer1","Deixou de ter alimentação saudável", "Não comeu suficiente", "Diminuida quantidade de comida", "Deixou de fazer uma refeição", "Sentiu fome2", "Dia inteiro sem comer2"]
d09.dropna(inplace=True)

```

##### O estudo foi baseado em duas categorias. A primeira se diz respeito a preocupação dos moradores: se os moradores tiveram a preocupação de que os alimentos acabassem antes de poderem comprar ou receber mais comida. A segunda se trata de uma certa "sofrência" por parte deles: os alimentos acabaram antes que os moradores tivessem dinheiro para comprar mais comida; os moradores ficaram sem dinheiro para ter uma alimentação saudável e variada; os moradores comeram apenas alguns alimentos que ainda tinham porque o dinheiro acabou. Se em algum domicílio ocorreu ao menos uma dessas "sofrências", ele se incorpora a variável 'SOFREU' da tabela.

```python
d13.loc[(d13["Alimentos acabaram"]=='Sim') | (d13["Sem dinheiro para uma alimentação saudável"]=='Sim') | (d13["Comeram pouco pois dinheiro acabou"]=='Sim'),'SOFREU'] = 'Sim'
d13.loc[(d13["Alimentos acabaram"]=='Não') & (d13["Sem dinheiro para uma alimentação saudável"]=='Não') & (d13["Comeram pouco pois dinheiro acabou"]=='Não'),'SOFREU'] = 'Não'
```

```python
d09.loc[(d09["Alimentos acabaram"]=='Sim') | (d09["Sem dinheiro para uma alimentação saudável"]=='Sim') | (d09["Comeram pouco pois dinheiro acabou"]=='Sim'),'SOFREU'] = 'Sim'
d09.loc[(d09["Alimentos acabaram"]=='Não') & (d09["Sem dinheiro para uma alimentação saudável"]=='Não') & (d09["Comeram pouco pois dinheiro acabou"]=='Não'),'SOFREU'] = 'Não'
```

```python
#numero de preocupacoes
print('2009 Frequências absolutas por Preocupação :')
ut2 = d09["Preocupação"].value_counts(sort=False).reindex(['Sim', 'Não'])
print(ut2,'\n')

print('2013 Frequências absolutas por Preocupação :')
ut1 = d13["Preocupação"].value_counts(sort=False).reindex(['Sim', 'Não'])
print(ut1,'\n')
```

##### Acima é possível observar as frequências absolutas de preocupação nos diferentes anos. Em 2009 ocorreram, em número, mais preocupações do que em 2013, o que podemos considerar (em um primeiro momento) como uma melhora na segurança alimentar dos domicílios. Porém, deve-se ter em conta que as tabelas filtradas dos dois anos, fez com que o número total a ser estudado de 2009 é de 25641 enquanto o de 2013 é 17490. Logo, os números tendem a ser menores nos gráficos e tabelas de 2013. Isso se deve ao fato da tabela ser filtrada a ponto de excluir todas as linhas onde o NaN aparecia, ou seja, se em algum momento a pergunta não era aplicável, ela foi excluída da tabela, a ponto de deixar apenas os domicílios em que os moradores responderam "sim" ou "não" para todas as perguntas da análise em questão. 

```python
#porcentagem de preocupacao
print('2009 Frequências relativas por Preocupação:')
ut2 = (d09["Preocupação"].value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut2,'\n')


print('2013 Frequências relativas por Preocupação:')
ut1 = (d13["Preocupação"].value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut1,'\n')
```

```python
#grafico de porcentagem de preocupacao
fig = plt.figure(figsize=(10, 5))

plt.subplot(121)
plot = ut2.plot(kind='pie',title='2009 Preocupação (%)',autopct='%.1f')

plt.subplot(122)
plot = ut1.plot(kind='pie',title='2013 Preocupação (%)',autopct='%.1f')
```

##### Aqui podemos observar que no total de pessoas sendo avaliadas a cada ano (tabela já filtrada), a porcentagem de "Preocupados" é praticamente a mesma nos dois anos. Mesmo a porcentagem sendo a mesma, o número absoluto foi maior em 2009, provavelmente ao fato do total de avaliados ser maior em 2009.

##### Abaixo vamos observar uma tabela cruzada entre preocupação e sofrência. Ou seja, daqueles que se preocuparam, quantos realmente sofreram ou não, e daqueles que não se preocuparam, quantos sofreram ou não. A primeira tabela mostra o número absoluto e a segunda a porcentagem.

```python
#2009 crosstab de preocupacao x "sofrencia" 
ct1=pd.crosstab(d09.SOFREU, d09["Preocupação"], margins=True, rownames=['Sofreu'], colnames=['2009 - Preocupação'])
ct1=ct1.loc[['Sim','Não','All']]
ct1
```

```python
#porcentagem 2009
ct3=(pd.crosstab(d09.SOFREU, d09["Preocupação"], margins=True, rownames=['Sofreu'], colnames=['2009 - Preocupação'], normalize='columns')*100).round(decimals=0)
ct3=ct3.loc[['Sim','Não']]
ct3
```

```python
#2013 crosstab de preocupacao x "sofrencia"
ct=pd.crosstab(d13.SOFREU, d13["Preocupação"], margins=True, rownames=['Sofreu'], colnames=['2013 - Preocupação'])
ct=ct.loc[['Sim','Não','All']]
ct
```

```python
#porcentagem 2013
ct2=(pd.crosstab(d13.SOFREU, d13["Preocupação"], margins=True, rownames=['Sofreu'], colnames=['2013 - Preocupação'], normalize='columns')*100).round(decimals=0)
ct2=ct2.loc[['Sim','Não']]
ct2
```

```python
plot = ct3.T.plot(kind='bar', stacked=True,title='2009 - Sofrência X Preocupação')
plot.set_ylim([0, 100])
plt.ylabel('Sofrência (em %)')

#Para legenda ficar fora do gráfico
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plot = ct2.T.plot(kind='bar', stacked=True,title='2013 - Sofrência X Preocupação')
plot.set_ylim([0, 100])
plt.ylabel('Sofrência (em %)')

#Para legenda ficar fora do gráfico
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

##### Podemos concluir aqui que a porcentagem comparativa mudou levemente de um ano para o outro. Em 2009,  80% das pessoas que se preocuparam, de fato sofreram. Já em 2013, o número caiu para 78%.  Enquanto 20% sim se preocupou porém não sofreu (em 2009) e 22% em 2013. Interessante observar que em ambos os anos, 0% das pessoas não se preocuparam E não sofreram. Ou seja, faltou certa preocupação de 5344 pessoas (somando os dois anos).

##### A partir de agora, vamos inserir o termo "consequência". Ele indica que se algum maior de idade em algum momento deixou de fazer alguma refeição porque não havia dinheiro para comprar comida; comeu menos do que achou que devia porque não havia dinheiro para comprar comida; sentiu fome mas não comeu porque não havia dinheiro para comprar comida; ou fez uma refeição ou ficou um dia inteiro sem comer porque não havia dinheiro para comprar comida então ele sofreu uma consequência.
