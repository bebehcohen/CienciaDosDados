```python
d13=dados13.loc[:,['V0105', 'V2103', 'V2105', 'V2107', 'V2109', 'V2113', 'V2115', 'V2117', 'V2121', 'V2124','V2126', 'V2128', 'V2130', 'V2132', 'V2134']]
d13.columns=["Total Moradores", "Preocupação", "Alimentos acabaram", "Sem dinheiro para uma alimentação saudável", "Comeram pouco pois dinheiro acabou", "Deixou de comer","Comeu menos", "Sentiu fome1", "Dia inteiro sem comer1","Deixou de ter alimentação saudável", "Não comeu suficiente", "Diminuida quantidade de comida", "Deixou de fazer uma refeição", "Sentiu fome2", "Dia inteiro sem comer2"]
d13.dropna(inplace=True)
```

```python
#numero de preocupacoes
print('Frequências absolutas por Preocupação:')
ut1 = d13["Preocupação"].value_counts(sort=False).reindex(['Sim', 'Não'])
print(ut1,'\n')
```

```python
#porcentagem de preocupacao
print('Frequências relativas por Preocupação:')
ut1 = (d13["Preocupação"].value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut1,'\n')
```

```python
#grafico de porcentagem de preocupacao
fig = plt.figure(figsize=(5, 5))
plt.subplot()
plot = ut1.plot(kind='pie',title='Preocupação (%)',autopct='%.1f')
```

```python
#crosstab de preocupacao x "sofrencia"
ct=pd.crosstab(d13.SOFREU, d13["Preocupação"], margins=True, rownames=['Sofreu'], colnames=['Preocupação'])
ct=ct.loc[['Sim','Não','All']]
ct
```

```python
print('Tabela Cruzada por total de COLUNAS (em %):')
ct2=(pd.crosstab(d13.SOFREU, d13["Preocupação"], margins=True, rownames=['Sofreu'], colnames=['Preocupação'], normalize='columns')*100).round(decimals=0)
ct2
```

```python
plot = ct2.T.plot(kind='bar', stacked=True,title='Sofrência X Preocupação')
plot.set_ylim([0, 100])
plt.ylabel('Sofrência (em %)')

#Para legenda ficar fora do gráfico
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
